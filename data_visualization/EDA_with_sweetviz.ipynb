```
import pandas as pd
import sweetviz
```

```
train= pd.read_csv('https://raw.githubusercontent.com/SrikanthVelpuri/House-Prices-Advanced-Regression-Techniques/master/train.csv')
```

```
test = pd.read_csv('https://raw.githubusercontent.com/SrikanthVelpuri/House-Prices-Advanced-Regression-Techniques/master/test.csv')
```

```
train.head()
```

```
my_report = sweetviz.analyze([train, "Train"], "SalePrice")
```

```
my_report.show_html('train_report.html')
```

```
my_report2 = sweetviz.compare([train, "Train"], [test, "Test"], "SalePrice")
```

```
my_report2.show_html('cmp_report.html')
```
